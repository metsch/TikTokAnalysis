```python
import pandas as pd;
import json

filename = "trump_1811_1604410542"
df = pd.read_csv("../data/"+filename+".csv")
display(df.head())
display(df.columns)
```

```python
n = len(df)
print(n)
error_count = 0
```

```python
df["itemInfos"][0]
```

```python
error_count = 0
def transform_row(row):   
    authorInfos_dict = row.authorInfos.replace("\'","\"")
    authorStats_dict = row.authorStats.replace("\'","\"")
    itemInfos_dict = row.itemInfos.replace("\'","\"")
    
    try:
        authorInfos = pd.DataFrame.from_dict(eval(authorInfos_dict),orient="index")
        authorStats = pd.DataFrame.from_dict(eval(authorStats_dict),orient="index")
        itemInfos = pd.DataFrame.from_dict(eval(itemInfos_dict),orient="index")
    except Exception:
        global error_count
        error_count = error_count + 1
        return
    
    
    authorInfos = authorInfos.transpose()
    authorInfos = authorInfos.filter(["userId","uniqueId","nickName","verified"],axis=1)
       
    authorStats = authorStats.transpose()
    authorStats = authorStats.filter(["followingCount","followerCount","heartCount","videoCount"],axis=1)
 
    itemInfos = itemInfos.transpose()
    itemInfos = itemInfos.filter(["id","authorId","shareCount","playCount","commentCount"],axis=1)


    textExtra = row.textExtra
    res = eval(textExtra)
    hashtags = set()
    
    for i in range(0,len(res)):
        if(res[i]["HashtagName"]):
            hashtags.add(res[i]["HashtagName"])
    
    row = pd.concat([
        itemInfos,
        authorStats,
        authorInfos
    ],axis=1)

    row["hashtags"] = None
    row.at[0,"hashtags"] = list(hashtags)
    row = row.rename(columns={"id":"itemId"})

    return row
```

```python
%%time
temp = transform_row(df.iloc[0])
data = pd.DataFrame(data=None,columns=temp.columns)

appended = []
for i in range(0,n-1):
    d = transform_row(df.iloc[i])
    appended.append(d)
    
data = pd.concat(appended)
print(len(data))
data = data.reset_index(drop=True)
```
